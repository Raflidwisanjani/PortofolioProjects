```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd
pd.set_option('display.max_rows', 99999)

import smtplib
```

```python
#connect to website

URL = 'https://www.maternaldisaster.com/shop/'

page = requests.get(URL)

soup1 = BeautifulSoup(page.content, 'html.parser')

soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')


print(soup2)
```

```python
product_list = []
products = soup2.find_all('div',class_="product-description text-center text-md-right")
for product in products:
  product_name = product.find('h3').text
  product_name = product_name.replace('\n ','')
  product_price = product.find('span',class_='price').text
  product_price = product_price.replace('Rp','').replace('Out of stock','').strip()
  product_list.append({'name':product_name,'price':product_price})
```

```python
df = pd.DataFrame(product_list)

df
```
